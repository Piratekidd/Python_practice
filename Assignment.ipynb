# Trend Analysis over month

```python
sample_data
```

```python
april_avg = sample_data['April_MT'].mean()
may_avg = sample_data['May_MT'].mean()
june_avg = sample_data['June_MT'].mean()
july_avg = sample_data['July_MT'].mean()
aug_avg = sample_data['August_MT'].mean()
sept_avg = sample_data['September_MT'].mean()
```

```python
july_avg/june_avg
```

```python
months = ['April', 'May', 'June', 'July', 'August', 'September']
math_scores = [april_avg, may_avg, june_avg, july_avg, aug_avg, sept_avg]

plt.figure(figsize=(8, 6))
plt.plot(months, math_scores, marker='o', color='b', label='Math Scores')

plt.title('Monthly Math Test Scores')
plt.xlabel('Months')
plt.ylabel('Scores')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# Model

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming 'months' and 'math_scores' are lists or 1D arrays
months = np.array(months).reshape(-1, 1)  # Convert to 2D array
math_scores = np.array(math_scores)  # Ensure math_scores is a 1D array

# One-hot encode the month names
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
months_encoded = encoder.fit_transform(np.array(months).reshape(-1, 1))

# Create and fit the model
model = LinearRegression()
model.fit(months_encoded, math_scores)

# Encode future months
future_months = ['October', 'November']  # Example future months
future_months_encoded = encoder.transform(np.array(future_months).reshape(-1, 1))

# Predict scores for future months
predicted_scores = model.predict(future_months_encoded)
print(predicted_scores)
```
